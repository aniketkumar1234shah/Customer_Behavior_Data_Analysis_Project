```python
import pandas as pd
from sqlalchemy import create_engine

df = pd.read_csv("customer_shopping.csv")


df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')


df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'}, inplace=True)

labels = ['young_adult','young','middle_aged','senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)


df.drop('promo_code_used', axis=1, inplace=True)



engine = create_engine(
    "postgresql+psycopg2://postgres:jiyajain@localhost:5432/customer_behaviour"
)


df.to_sql("customer_cleaned", engine, if_exists="replace", index=False)

print("✅ Cleaned table uploaded successfully!")
```
